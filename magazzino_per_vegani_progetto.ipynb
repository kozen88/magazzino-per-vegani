```python
def is_product_correct(string):
    """
    Controlla se la stringa inserita contiene solo lettere o spazi.

    Parametri
    - string (str): la stringa da controllare.
    """
    # pattern to specify only letters or space
    pattern = re.compile(r'^[a-zA-Z ]+$')
    
    return bool(pattern.match(string))

def product_check_insertion(sell=False):
    """
    Controlla gli input inserit dall'utente sul prodotto.
    
    Return
    - (tuple): tupla contenente le informazioni valide da inserire.
    """
    # check on product name 
    while True:
        prod = input("Nome del prodotto: ").lower()
        try:
            assert is_product_correct(prod), "L'input deve contenere solo lettere e spazi vuoti."
        except AssertionError as e:
            print(e)
            continue
        else:
            print("Input valido.")
            break
    
    # check on positive and correct quantity
    while True:
        quantity = input("Quantità: ")
        try:
            quantity = int(quantity)  
            if quantity > 0:
                break  # Exit cicle of input
            else:
                print("La quantità deve essere un numero strettamente maggiore di zero. Riprova.")
        except ValueError:
            print("Input non valido. La quantità deve essere un numero intero. Riprova.")

    if sell:
        return (prod, quantity)
    else:
        # check on purchase_price
        while True:
            purchase_price = input("Prezzo di acquisto: ")
            try:
                purchase_price = float(purchase_price)  
                if purchase_price > 0:
                    break  # Exit cicle of input
                else:
                    print("Il prezzo di acquisto deve essere un numero strettamente maggiore di zero. Riprova.")
            except ValueError:
                print("Input non valido. Il prezzo di acquisto deve essere un numero. Riprova.")
            
        # check on selling_price
        while True:
            selling_price = input("Prezzo di vendità: ")
            try:
                selling_price = float(selling_price) 
                if selling_price > 0:
                    break  # Exit cicle of input
                else:
                    print("Il prezzo di vendita deve essere un numero strettamente maggiore di zero. Riprova.")
            except ValueError:
                print("Input non valido. Il prezzo di vendita deve essere un numero. Riprova.")
              
        return (prod, quantity, purchase_price, selling_price)

    
def add_product(store=STORE):
    """
    La funzione permette di aggiungere un prodotto al magazzino.

    Parametri:
    - store (str) [default=STORE]: nome del file json che contiene il magazzino.
    """

    if not os.path.exists(store):

        init_store(store)
        print("Adesso è possibile aggiungere il tuo primo prodotto al magazzino.")

    
    product, quantity, purchase_price, selling_price = product_check_insertion()

    dict_store = load_store(store)

    # print(dict_store)
    if product in dict_store:
        # product exists
        print(f"Il prodotto {product} che si vuole aggiungere è presente in magazzino.")
        
        choice = input("Oltre alla quantità si desiderà modificare il prezzo con i dati forniti? [si/NO]:").lower()
        
        while choice!='si' and choice!='no':
            choice = input("\nLa scelta effettuata non è correta. Digità [si/no]" +
                             "\nOltre alla quantità si desiderà modificare il prezzo con i dati forniti? [si/NO]:").lower()
            
        if choice == 'si':
            dict_store[product]['quantity'] += quantity
            dict_store[product]['purchase_price'] = purchase_price
            dict_store[product]['selling_price'] = selling_price
            
        elif choice == 'no':
            dict_store[product]['quantity'] += quantity
            
    else:
        # product do not exists
        dict_store[product] = {}
        dict_store[product]['quantity'] = quantity
        dict_store[product]['purchase_price'] = purchase_price
        dict_store[product]['selling_price'] = selling_price

    save_store(dict_store, store)
    print(f"AGGIUNTO: {quantity} x {product}")
```

```python
def get_product_from_selling_note(note=SELLING_BOOK):
    """
    Questa funzione legge il file csv delle vendite note e recupera una lista contenente i nomi dei prodotti presenti
    presi una singola volta, ovvero senza ripetizioni.

    Parametri:
    - note (str) [default=SELLING_BOOK]: nome del file csv contenente le vendite registrate.

    Return:
    - products_list (list): lista di stringhe rappresentanti i nomi dei prodotti.
    """
    
    products_set = set()  
    
    with open(note, mode='r', newline='') as file:
        reader = csv.reader(file)
        
        header = next(reader)
        
        # check colon PRODUCT 
        if "PRODUCT" in header:
            for row in reader:
                # get name of the product and add to the list
                product_name = row[header.index("PRODUCT")]  
                products_set.add(product_name)  
    
    products_list = list(products_set)
    
    return products_list

def load_selling_dictionary(note=SELLING_BOOK):
    """
    Questa funzione andrà a caricare in memoria un dizionario delle vendite ove ogni
    prodotto rappresenta la chiave per accedere ad un sotto dizionario contenete le chiavi
    quantity per la quantità di prodotto venduta e total_price per il totale ricavato come lordo.

    Parametri:
    - note (str) [default=SELLING_BOOK]: nome del file csv che rappresenta il registro delle vendite.

    Return:
    - product_dict (dict): il dizionario contenente il recap delle info sulle vendite per prodotto.
    """
    
    product_dict = {}
    
    # get unique product's name
    products_list = get_product_from_selling_note(note)
    
    # init product_dict
    for product_name in products_list:
        product_dict[product_name] = {'quantity': 0, 'total_price': 0.0}
    
    with open(note, mode='r', newline='') as file:
        reader = csv.reader(file)
        
        header = next(reader)
        
        # check on header
        if "PRODUCT" in header:
            product_index = header.index("PRODUCT")
            quantity_index = header.index("QUANTITY")
            total_price_index = header.index("TOTAL_PRICE")
            
            # read each line and update the product_dict
            for row in reader:
                product_name = row[product_index]
                
                quantity = int(row[quantity_index])
                product_dict[product_name]['quantity'] += quantity
                
                total_price = float(row[total_price_index])
                product_dict[product_name]['total_price'] += total_price
    
    return product_dict


def is_selling_possible(dict_prod, name_prod, quantity):
    """
    Questa funzione controlla che il prodotto registrato per la vendita sia disponibile.

    Parametri:
    - dict_prod (dict): dizionario contenente i prodotti e le informazioni.
    - name_prod (str): nome del prodotto che si vuole vendere.
    - quantity (int): numero che identifica la quantità del prodotto richiesta.

    Return:
    - (product_exists, selling) (tuple:boolean, boolean): indica rispettivamente che il prodotto esiste e che è vendibile.
    """
    
    product_exists = False
    selling = False
    
    if name_prod in dict_prod:
        product_exists = True
        
        if dict_prod[name_prod]['quantity'] >= quantity:
            # product exists and selling OK
            selling = True
            return (product_exists, selling)
            
        else:
            # product exists, but not enough for selling
            return (product_exists, selling)
    else:
        # product do not exists and impossible selling
        return (product_exists, selling)

def insert_item_to_sell():
    """
    Questa funzione permette di immettere tutti i prodotti che si vogliono vendere con le rispettive quantità, controllando 
    la correttezza delle informazioni. Per ogni prodotto viene creata una tupla (prodotto, quantità) la quale viene aggiunta
    in una lista che rappresenta tutti i prodotti che si vogliono vendere la quale sarà ritornata dal sistema. 

    Retrun
    - list_item_to_sell  (list of tuple): contine le tuple (prodotto, quantità)
    """
    list_item_to_sell = []
    
    while True:
        product_to_sell, number =  product_check_insertion(sell=True)
        item_to_sell = (product_to_sell, number)
        list_item_to_sell.append(item_to_sell)
        
        choice = input("Aggiungere un altro prodotto? [si/no]: ").lower()
        
        while choice != "si" and choice != "no":
            choice = input("\nLa scelta effettuata non è correta. Digità [si/no]\nAggiungere un altro prodotto? [si/no]:").lower()
            
        if choice == "no":
             break

    return list_item_to_sell

def update_selling_book(data, selling_book_just_created, note=SELLING_BOOK):
     """
     Questa funziona va ad aggiornare le vendite nel registro vendite csv.

     Parametri:
     - data (tuple): contiene i dati usati con i quali aggiornare il csv
     - selling_book_just_created (bool): usata per creare ex novo il registro delle vendite.
     - note (str) [default=SELLING_BOOK]: nome del file csv che rappresenta il registro delle vendite.

     Return:
     - record_of_sell (str): una stringa che contiene le informazioni di vendita.
     """
     with open(note, mode='a', newline='') as file:
        writer = csv.writer(file)
        
        # when the file do not exists write the header
        if not selling_book_just_created:
            writer.writerow(['PRODUCT', 'QUANTITY', 'TOTAL_PRICE'])
        
        # write data in csv
        writer.writerow((data[0], data[1], data[3]))

        record_of_sell = f"{data[1]} x {data[0]} each at €{data[2]}"

     return record_of_sell


def sell_product(product_to_sell, number, store=STORE, note=SELLING_BOOK):
    """
    La funzione permette di effettuare la vendita di un prodotto previo controllo nel magazzino.
    Le informazioni inerenti la vendita saranno scritte in un opportuno file csv.

    Parametri
    - product_to_sell (str): nome del prodotto da vendere.
    - number (int): quantità del prodotto che si vuole vendere.
    - store (str) [default=STORE]: nome del file json che contiene i dati dei prodotti in magazzino.
    - note (str) [default=SELLING_BOOK]: nome del file csv che registra le vendite effettuate.

    Return:
    - (record_of_sell, tot_price) (tuple): tupla che contiene la stringa con le informazioni di vendita e il prezzo totale per il prodotto.
    """
    tot_price = 0
    record_of_sell = ""

    # the store exists and works
    if os.path.exists(store):
        # load products in ram
        dict_store = load_store(store)

        # check availability of product 
        product_is_present, selling_is_ok = is_selling_possible(dict_store, product_to_sell, number)
        # debug prints
        # print(f"Il prodotto è presente? {product_is_present}")
        # print(f"La vendita è possibile{selling_is_ok}")
        
        if product_is_present and selling_is_ok:
            dict_store[product_to_sell]['quantity'] -= number
            sell_price = dict_store[product_to_sell]['selling_price']
            tot_price = sell_price * number
            
            # store update
            save_store(dict_store, store)
            
            # check if file exists
            file_exists = os.path.exists(note)

            # update selling book
            record_data = (product_to_sell, number, sell_price, tot_price)
            record_of_sell = update_selling_book(record_data, file_exists, note=SELLING_BOOK)
    
        elif product_is_present and (not selling_is_ok):
            print(f"La vendita di {product_to_sell} non è possibile, quantità in magazzino insufficiente.")
            
        elif dict_store and (not product_is_present):
            if dict_store.keys():
                print(f"La vendità di {product_to_sell} non è possibile, prodotto non disponibile.!")
                
            else:
                print("ATTENZIONE! il magazzino è vuoto! Vendita impossibile!")
    else:
        # se non abbiamo a disposizine il magazzino
        print("ATTENZIONE! al momento non esiste alcun magazzino! Vendita impoossibile!")

    return (record_of_sell, tot_price)
```

```python
def calculate_gross_profit(data):
    """
    Calcola il guadagno lordo delle vendite effettuate.

    Parametri:
    - data (dict): contiene le informazioni di vendita del registro in forma di dizionario.

    Return:
    - gross_profit (float): il guadagno lordo calcolato.
    """
    
    gross_profit = 0
    
    for key in data.keys():
        gross_profit += data[key]['total_price']

    return gross_profit

def calculate_net_profit(sell_data, store_data):
    """
    Calcola il guadagno netto delle vendite effettuate.

    Parametri:
    - sell_data (dict): contiene le informazioni del registro di vendita sotto forma di dizionario.
    - store_data (dict): contiene il magazzino caricato in memoria sotto forma di dizionario.

    Return
    - net_profit (float): il guadagno netto delle vendite.
    """
    
    net_profits = 0

    for key in sell_data.keys():
        net_profits += sell_data[key]['quantity'] * store_data[key]['purchase_price']

    
    return net_profits

def show_profits(store=STORE, note=SELLING_BOOK):
    """
    Questa funzione permette di recuperare i profitti lordo e netto e di mostrali.

    Parametri:
    - store (str) [default=STORE]: il nome del file json contenente il magazzino.
    - note (str) [default=SELLING_BOOK]: il nome del file csv contenente il registro vendita.
    """
    
    if os.path.exists(note) and os.path.exists(store):
    
        selling_dict = load_selling_dictionary(note)
        store_dict = load_store(store)
    
        gross = calculate_gross_profit(selling_dict)
        net = calculate_net_profit(selling_dict, store_dict)
    
        print(f"Profitto: lordo=€{gross:.2f} netto=€{net:.2f}")
    else:
        print(f"Non è possibile mostrare i profitti file {note} e/o {store} mancanti.")
```
